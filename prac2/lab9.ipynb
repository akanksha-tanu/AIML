```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

tou=0.5
data=pd.read_csv('../lab9.csv')
xtrain=np.array(data.total_bill)
xtrain=xtrain[:,np.newaxis]
ytrain=np.array(data.tip)
xtest=np.array([i/10 for i in range(500)])
xtest=xtest[:,np.newaxis]
ytest=[]

for r in range(len(xtest)):
    wts=np.exp(-np.sum((xtrain-xtest[r])**2,axis=1)/2*tou**2)
    w=np.diag(wts)
    factor1=np.linalg.inv(xtrain.T.dot(w).dot(xtrain))
    parameters=factor1.dot(xtrain.T).dot(w).dot(ytrain)
    prediction=xtest[r].dot(parameters)
    ytest.append(prediction)

plt.plot(xtrain.squeeze(),ytrain,'o')
plt.plot(xtest.squeeze(),ytest,'o')
plt.show()
```
